**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical 3&times;3 Prewitt filters:

```python
def prewitt_h(img_in):
    result = numpy.zeros(img_in.shape)
    size = 3
    n = (size - 1) // 2
    for p in numpy.ndindex(img_in.shape):
        neighborhood_sum = 0
        if p[0]< img_in.shape[0]-n and p[0]>=n and p[1]>=n and p[1]< img_in.shape[1]-n:
            for q in numpy.ndindex(size,size):
                y= q[0]+p[0]-n
                x= q[1]+p[1]-n
                if q[1] == size: #weil convolution -> wurde geflippt
                    neighborhood_sum += (img_in[y,x]*1)
                if q[1]>0 and q[1]<size:
                    neighborhood_sum += (img_in[y,x]*0)
                if q[1]==0:
                    neighborhood_sum += (img_in[y,x]*(-1))
            result[p] = neighborhood_sum/(size*2)
    return result
    
def prewitt_v(img_in):
    result = numpy.zeros(img_in.shape)
    size = 3
    n = (size - 1) // 2
    for p in numpy.ndindex(img_in.shape):
        neighborhood_sum = 0
        if p[0]< img_in.shape[0]-n and p[0]>=n and p[1]>=n and p[1]< img_in.shape[1]-n:
            for q in numpy.ndindex(size,size):
                y= q[0]+p[0]-n
                x= q[1]+p[1]-n
                if q[0] == size: #weil convolution -> wurde geflippt
                    neighborhood_sum += (img_in[y,x]*1)
                if q[0]>0 and q[1]<size:
                    neighborhood_sum += (img_in[y,x]*0)
                if q[0]==0:
                    neighborhood_sum += (img_in[y,x]*(-1))
            result[p] = neighborhood_sum/(size*2)
    return result
```

**<a id="#">Task 4.1.3.</a>** Test your above solution using the functions `prewitt_h` and `prewitt_v` for the previously loaded image and show the result:

```python
plt.figure()
plt.imshow(prewitt_h(ast))
plt.colorbar()

plt.figure()
plt.imshow(prewitt_v(ast))
plt.colorbar()
```

```python
import skimage.io
assert numpy.allclose(prewitt_h(ast), skimage.io.imread('data/astronaut_prewitt_h.tiff'), atol = 1/255)
assert numpy.allclose(prewitt_v(ast), skimage.io.imread('data/astronaut_prewitt_v.tiff'), atol = 1/255)
```
